```python
import pandas as pd
import numpy as np
```

```python
trans_data = pd.read_csv('C:\\Kaggle competetion datasets\\Elo Merchent Category Recommendation\\historical_transactions.csv',
                        usecols=['card_id','merchant_category_id', 'purchase_amount', 'city_id', 'state_id', 'subsector_id'])
```

# User Defined Functions

```python
########  Calculate No Of Transactions  ##########

def num_of_trans(df, col1,col2,col3):
    data = df.groupby([col1,col2])[col3].size().reset_index()
    data.columns = [col1, col2, 'No_of_transactions']
    return(data)
```

```python
########  Calculate Amount Of Transactions  ##########

def amount_of_trnas(df1, col_1, col_2, col_3):
    data1 = df1.groupby([col_1, col_2]).apply(lambda x: np.sum(x[col_3])).reset_index()
    data1.columns = [col_1, col_2, 'total_amount_of_purchase']
    return(data1)
```

```python
########  Pivot Table  ##########

def pivot_df(df, col_a, col_b, col_c):
    data_pivot = df.pivot(index = col_a, columns = col_b, values = col_c)
    return(data_pivot)
```

# 1. Capture number of transaction 

## No Of Transactions Of Customer Per Each Merchant Category

```python
total_trans = num_of_trans(trans_data, 'card_id', 'merchant_category_id', 'purchase_amount')
total_trans.tail(7)
```

```python
total_trans_pivot = pivot_df(total_trans, 'card_id', 'merchant_category_id', 'No_of_transactions')
total_trans_pivot.tail(7)
```
